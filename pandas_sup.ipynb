✅ Step 7 — Remove Duplicate Rows

```
print("Before:", df.shape[0])

df = df.drop_duplicates()

print("After:", df.shape[0])
```

✅ Step 8 — Convert Date Column

```
df['Date'] = pd.to_datetime(df['Date'])
```

```
df['Month'] = df['Date'].dt.month
```

```
total_sales = df['Total'].sum()
print("Total Sales:", total_sales)
```

✅ With Checking (Recommended Steps)

```
import pandas as pd

df = pd.read_csv("supermarket_sales_500_rows.csv")

# Before removing duplicates
print("Total rows before:", df.shape[0])
print("Duplicate rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

# After removing duplicates
print("Total rows after:", df.shape[0])
```

✅ Remove Duplicates Based on Specific Column

```
df = df.drop_duplicates(subset=['Invoice ID'])
```

✅ Keep Last Duplicate Instead of First

```
df = df.drop_duplicates(keep='last')
```

✅ Save Cleaned Data

```
df.to_csv("cleaned_data.csv", index=False)
```
