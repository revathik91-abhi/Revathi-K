✅ Step 1 — Import Libraries

```python
import pandas as pd
import numpy as np
```

✅ Step 2 — Load the Dataset

```python
df = pd.read_csv("/content/supermarket_sales_500_rows (1).csv")
```

✅ Step 3 — Convert DataFrame to NumPy Array

```python
arr = df.to_numpy()
```

✅ Step 4 — Remove Duplicate Rows Using NumPy

```python
arr= arr.astype(str)
unique_arr = np.unique(arr, axis=0)
```

✅ Step 5 — Convert NumPy Array Back to DataFrame

```python
df_clean = pd.DataFrame(unique_arr, columns=df.columns)
```

✅ Step 6 — Check Number of Rows After Removing Duplicates

```python
print("Rows after removing duplicates:", df_clean.shape[0])
```

✅ Step 7 — Save the Cleaned Dataset

```python
df_clean.to_csv("cleaned_numpy_data.csv", index=False)
```

✅ Step 8 — Load Dataset

```python
df.head()
df.shape
df.info()
```

✅ Step 9 — Check Missing Values

```python
df.isnull().sum()
```

✅ Step 10 — Handle Missing Values

Fill numeric columns with mean

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

Fill categorical columns with mode

```python
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

✅ Step 11 — Remove Duplicate Values

```python
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
```

✅ Step 12 — Convert Data Types

```python
df['Date'] = pd.to_datetime(df['Date'])
```

✅ Step 13 — Feature Engineering (Create New Columns)


Extract Month

```python
df['Month'] = df['Date'].dt.month
```

Extract Day

```python
df['Day'] = df['Date'].dt.day
```

✅ Step 14 — Rename Columns

```python
df.rename(columns={'Product line': 'Product_Line'}, inplace=True)
```

✅ Step 15 — Handle Outliers

```python
Q1 = df['Total'].quantile(0.25)
Q3 = df['Total'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Total'] >= Q1 - 1.5*IQR) & (df['Total'] <= Q3 + 1.5*IQR)]
```

✅ Step 16 — Data Formatting

```python
df.columns = df.columns.str.lower()
```

✅ Step 17 — Save Cleaned Data

```python
df.to_csv("cleaned_supermarket_sales.csv", index=False)
```
